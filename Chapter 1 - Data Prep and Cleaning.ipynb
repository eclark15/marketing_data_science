### Using MERGE and JOINS

```python
# create our mini DFs
df3 = pd.DataFrame() 
data = [['Sushmita',31.26605567],['Adam',17.90052851],['Benny',265.2360861],['Anurag',42.47069602]]
df3 = pd.DataFrame(data, columns=['users', 'timeSpent'])
df3
```

```python
df4 = pd.DataFrame() 
data1 = ['Adam',20],['Anurag',],['Benny',15],['Sushmita',2],['Apoorva',7]
df4 = pd.DataFrame(data1, columns=['users', 'cost'])
df4
```

```python
# Merge = JOIN; JOIN df4 LEFT JOIN df4 ON Users column to create new DF5 
    # can use inner, outer, left and right JOINs
    
df5 = df3.merge(df4, on='users', how = 'left') #df3 is the main table, LEFT JOIN 
df5
```

```python
df5 = df5.fillna(0) # any NULLs fill with 0
df5
```

```python
# when the column you are JOINing by have different names 
df6 = pd.DataFrame() 
data2 = ['Adam',20],['Anurag',],['Benny',15],['Sushmita',2],['Apoorva',7]
df6 = pd.DataFrame(data1, columns=['users2', 'cost'])
df6.head()

df7 = df3.merge(df6, left_on='users', right_on='users2', how = 'left')
df7.head()
```

```python
df7 = df7.fillna(df7.mean()) #fill in the NAs with the mean of the column 
df7
```

### Using MAP Function to Rename Values -- Female = F

```python
# updating the column gender 
df7['Gender'] = ["Female", "Male", "Male", "Female"] # add a column to an exisitng DF 

df7['Gender'] = df7['Gender'].map({"Female":"F", "Male": "M", "Nonbinary":"N"}) # put the mapping into a dict
df7.head()
```
